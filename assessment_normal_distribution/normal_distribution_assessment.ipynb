# Normal Distribution Notebook

The normal distribution, also known as the Gaussian distribution, is a probability distribution that is symmetric about the mean, with its spread determined by the standard deviation. The normal distribution is defined by the following probability density function:

$f(x) = \frac{1}{\sqrt{2\pi\sigma^2}} e^{-\frac{(x-\mu)^2}{2\sigma^2}}$

where $\mu$ is the mean and $\sigma$ is the standard deviation.

***

### Applications of the Normal Distribution
#### The normal distribution has a wide range of applications, including:

Modeling data that is naturally symmetric, such as human heights or IQ scores. Predicting the probability of events occurring, such as the probability of a coin flip landing heads or tails. Modeling the behavior of stock prices or other financial data. Overall, the normal distribution is a powerful tool in statistical analysis and probability theory, and it has a wide range of applications in a variety of fields.

***

## Main Concepts

### Mean and Standard Deviation

The mean of a normal distribution is the center of the distribution, and the standard deviation determines the spread of the distribution. A larger standard deviation means that the data is more spread out, while a smaller standard deviation means that the data is more concentrated around the mean.

### Symmetry

The normal distribution is symmetric about the mean, which means that if you were to fold it along the y-axis, the two halves would match perfectly. This property makes the normal distribution very useful in statistical analysis, as it allows us to make predictions about the probability of events occurring.

### The 68-95-99.7 Rule

One of the most useful properties of the normal distribution is the 68-95-99.7 rule, which states that:

- Approximately 68% of the data lies within one standard deviation of the mean.
- Approximately 95% of the data lies within two standard deviations of the mean.
- Approximately 99.7% of the data lies within three standard deviations of the mean.

This rule is often used to make predictions about the probability of events occurring in a normally distributed data set.

### Standard Normal Distribution
The standard normal distribution is a normal distribution with a mean of 0 and a standard deviation of 1. It is often used as a reference distribution when performing statistical tests or making predictions about a normally distributed population.

### Z-Score
A z-score is a measure of how many standard deviations a data point is from the mean of a distribution. It is calculated as:

$$z = \frac{x - \mu}{\sigma}$$

where $x$ is the data point, $\mu$ is the mean, and $\sigma$ is the standard deviation.

The z-score can be used to determine the probability of a data point occurring in a normal distribution. For example, a z-score of 1.96 corresponds to a probability of 95% (since approximately 95% of the data lies within two standard deviations of the mean).

### Percentiles
Percentiles describe the value below which a certain percentage of the distribution falls. For example, the 50th percentile (also known as the median) is the value below which 50% of the distribution falls.

### Central Limit Theorem
The central limit theorem states that the distribution of the mean of a large number of independent, identically distributed random variables will be approximately normal, regardless of the underlying distribution of the variables. This theorem has important implications in statistical analysis, as it allows us to make predictions about the mean of a population based on a sample, provided that the sample is large enough.

## Visualizing the Normal Distribution

This is a histogram to visualize the normal distribution. In the following example, I generated 1000 random samples from a normal distribution with a mean of 0 and a standard deviation of 1, and then plotted the histogram of the samples:

```python
# Importing necessary libraries
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import scipy
import random
from scipy import stats
from sklearn.neighbors import KernelDensity
from scipy.stats import norm
```

```python
# Generate 1000 random samples from a normal distribution with a mean of 0 and a standard deviation of 1
samples = np.random.normal(0, 1, 10000)
fig, ax = plt.subplots(figsize=(14, 5))
# Plot the histogram of the samples
plt.hist(samples, bins=100, density=True)
plt.show()
```

### Visualizing different sigma values. Sigma stands for standard deviation the measure of the amount of variation or dispersion of a set of values.

```python
%matplotlib inline
# Set the range of x values to plot
x = np.linspace(-10, 10, 1000)

# Set the different values of sigma to use
sigmas = [0.5, 1, 1.5, 2]
fig, ax = plt.subplots(figsize=(15, 6))
# Plot the normal distribution function for each sigma value
colors = ['b', 'g', 'r', 'c']
for i, sigma in enumerate(sigmas):
    color = random.choice(colors)
    plt.plot(x, scipy.stats.norm.pdf(x, 0, sigma), label=rf'$\sigma = {sigma}$')
 
plt.legend()
plt.show()
```

## Q-Q Plot

A Q-Q plot (quantile-quantile plot) is a graphical tool used to assess whether two samples come from the same distribution. It plots the quantiles of one sample against the quantiles of the other sample, and deviations from the expected straight line relationship can indicate differences in the underlying distributions.

```python
# Create the Q-Q plot
fig = plt.figure()
fig, ax = plt.subplots(figsize=(15, 6))
res = stats.probplot(samples, plot=plt)
plt.show()
```

## Kernel Density Plot

A Kernel Density plot is a non-parametric way to estimate the probability density function of a random variable. We can use a Kernel Density plot to visualize the shape of the distribution of a set of data.

To create a kernel density plot in Python, we can use the seaborn library:

```python
# Create the kernel density plot
fig, ax = plt.subplots(figsize=(14, 5))
sns.kdeplot(samples)
plt.show()
```

## Violin Plot
A violin plot is a graphical representation of statistical data that combines a Kernel Density plot with a box plot. It is particularly useful for comparing the distribution of data between multiple groups.

To create a violin plot in Python, we can use the seaborn library:

```python
# Create the violin plot
fig, ax = plt.subplots(figsize=(14, 6))
sns.violinplot(data=samples, color="red")
plt.show()
```

## Scatter Plot
A scatter plot is a graphical representation of statistical data that displays the values of two variables plotted on a coordinate grid. We can use a scatter plot to visualize the relationship between two variables and identify patterns and trends.

```python
# Create the scatter plot
fig, ax = plt.subplots(figsize=(15, 6))
plt.scatter(range(10000), samples, color="orange")
plt.show()
```

## Line Plot
A line plot is a graphical representation of statistical data that displays the values of a variable over time. We can use a line plot to visualize the trend of a variable over time and identify patterns and trends.

```python
# Create the line plot
fig, ax = plt.subplots(figsize=(15, 6))
plt.plot(range(10000), samples, color="green")
plt.show()
```

## Kernel Density Estimation Plot
A kernel density estimation plot is a non-parametric way to estimate the probability density function of a random variable. We can use a Kernel Density estimation plot to visualize the shape of the distribution of a set of data.

To create a kernel density estimation plot in Python using sklearn, we can use the Kernel Density class:

```python
# Create the kernel density estimation plot
fig, ax = plt.subplots(figsize=(14, 5))
kde = KernelDensity(kernel='gaussian', bandwidth=0.5).fit(samples.reshape(-1, 1))
x_plot = np.linspace(-5, 5, 1000)
log_dens = kde.score_samples(x_plot.reshape(-1, 1))
plt.plot(x_plot, np.exp(log_dens))
plt.show()
```

## Symmetry plot
The plot of the bell curve for a normal distribution with a mean of 0 and a standard deviation of 1, along with a vertical line at the mean to show the symmetry of the distribution.

```python
# Set mean and standard deviation
mean = 0
std = 1

# Generate a range of x values from -5 to 5
x = np.linspace(-5, 5, 100)
fig, ax = plt.subplots(figsize=(14, 5))
# Use scipy's norm.pdf function to get the probability density function for the normal distribution
y = norm.pdf(x, mean, std)

# Plot the bell curve
plt.plot(x, y)

# Add a vertical line at the mean
plt.axvline(mean, color='r', linestyle='--')

# Add a title and labels for the x and y axes
plt.title("Symmetry of normal distribution")
plt.xlabel("X")
plt.ylabel("Probability density")

# Show the plot
plt.show()
```

## Swarm Plot
A swarm plot is a graphical representation of statistical data that displays all the data points in a set, rather than just summarizing them with a few statistical measures like a box plot. We can use a swarm plot to visualize the distribution of data and identify patterns and trends.

```python
# Create the swarm plot
fig, ax = plt.subplots(figsize=(14, 5))
sampless = np.random.normal(0, 1, 500)
sns.swarmplot(data=sampless, color="purple")
plt.show()
```
